# Data Cleaning 

```python
df.info()
```

```python
#checkin for null values

df.isna().sum()
```

```python
df.shape
```

```python
#Checking for duplicates

df.duplicated().sum()
```

```python
df["date"].value_counts()
```

```python
df["date"].isna().sum()
```

```python
df.dropna(subset = ["date"], inplace = True)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.sample(n=2)
```

```python
#converting the date column to a datetime

df["date"] = pd.to_datetime(df["date"])
```

```python
df.info()
```

```python
#converting the time column to a timedelta

df["time"] = pd.to_timedelta(df["time"])
```

```python
df.info()
```

```python
df.head()
```

```python
#Extracting the sales column from the quantity and price column

df["sales"] = df["quantity"] * df["price"]

df.head()
```

```python
#Extracting the month column from the date column

df["months"] = df["date"].dt.month_name()

df.head()
```

```python
#Extracting the week days column from the date column

df["week_days"] = df["date"].dt.day_name()

df.head()
```

```python
#Extracting the hours column from the time column

df["hour"] = df["time"].dt.components["hours"]

df.head()
```

```python
df.info()
```

```python
df["hour"].value_counts()
```

```python
#Extracting the time period column from the hour column

h = df["hour"]

df["time period"] = np.select([h.lt(12), h.between(12,16), h.ge(17)], ["Morning", "Afternoon", "Evening"])

df.head()
```

```python
df["time period"].value_counts()
```

```python
df.info()
```

# Data Exploration

```python
total_sales = df["sales"].sum()

print(f"The total sales is {total_sales:,.2f}")
```

```python
total_quantity_sold = df["quantity"].sum()

print(f"The total quantity sold was {total_quantity_sold:,}")
```

```python
total_orders = df["order_id"].count()

print(f"There was a total order of {total_orders:,}")
```

```python
pizza_type = df["pizza_type_id"].nunique()

print(f"There are {pizza_type} pizza type being sold")
```

```python
pizza_avg_price = df["price"].mean()

print(f"The average price for the pizza is{pizza_avg_price: .2f}")
```

```python
#Average price per pizza

avg_price = df.groupby("name")["price"].mean().reset_index()

avg_price.rename(columns = {"price":"avg_price"}, inplace = True)
```

```python
print("The average price for each pizza is shown below")

avg_price.sort_values(by = "avg_price", ascending = False).reset_index(drop=True)
```

```python
df.head()
```

```python
#hour of the day with most order

hour_with_most_ord = df.groupby("hour")["order_id"].count().reset_index()

hour_with_most_ord.sort_values(by="order_id", ascending=False).reset_index(drop=True)
```

12pm in the day has the most order and when you look through it, you notice there are more orders in the afternoon hours 

```python
#Time of day with the most sales

peak_time_period = df.groupby("time period")["sales"].sum().reset_index()

peak_time_period
```

This is the grouping for the hours of the day, grouped into Morning, Afternoon and Evening

```python
sns.countplot(data = df, x = "time period")
plt.title("Total Sales per Time Period")
plt.xlabel("Time Period")
plt.ylabel("Sales")
plt.show()
```

The above chart is a visual to show that there are more orders in the Afternoon, then the Evening while the morning has the least orders 

```python
#Weekly Trend

day_order = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]

peak_day = df.groupby("week_days")["sales"].sum().reset_index()

peak_day["week_days"] = pd.Categorical(peak_day["week_days"], categories = day_order, ordered = True)

peak_day = peak_day.sort_values(by = "week_days").reset_index(drop = True)

peak_day
```

```python
plt.plot(peak_day["week_days"], peak_day["sales"])
plt.xticks(rotation = 90)
plt.title("Weekday Trend")
plt.xlabel("Week Days")
plt.ylabel("Sales")

plt.show()
```

This shows the weekly trend and Friday has the highest sales 

```python
#The bestselling pizzas

top_pizza_sales = df.groupby("name")["order_id"].count().reset_index()

top_pizza_sales.sort_values(by = "order_id", ascending = False).head().reset_index(drop=True)
```

The above shows the top 5 best selling pizzas. The Classic Deluxe Pizza is the best selling pizza with a total order of 2,416 and The Barbecue Chicken Pizza as the second with a total order of 2,372. 

```python
#Monthly Trend

month_order = ["January","February","March","April","May","June","July",
               "August","September","October","November","December"]

peak_months = df.groupby("months")["sales"].sum().reset_index()

peak_months["months"] = pd.Categorical(peak_months["months"], categories = month_order, ordered = True)

peak_months = peak_months.sort_values(by = "months").reset_index(drop = True)

peak_months.sort_values(by="sales", ascending = False).reset_index(drop = True)
```

```python
plt.plot(peak_months["months"], peak_months["sales"])
plt.xticks(rotation = 90)
plt.title("Monthly Trend")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```

The above shows the monthly sales trend, July made the highest sales with a sum of 72,557.90 while May had the second highest sales with a sum of 71402.75

```python
#Pizzas with the lowest revenue

low_rev = df.groupby("name")["sales"].sum().reset_index()
low_rev.sort_values(by = "sales", ascending = True).head().reset_index(drop=True)
```

The above shows the least 5 performing pizza, The Brie Carre Pizza is the least performing pizza, generating the sum of 11,588.50 in revenue  

```python
#Pizza with the highest revenue 

top_rev = df.groupby("name")["sales"].sum().reset_index()
top_rev.sort_values(by = "sales", ascending = False).head().reset_index(drop=True)
```

The above shows the top performing pizzas in revenue generation

```python
#Pizza category with the highest revenue

pizza_cat = df.groupby("category")["sales"].sum().reset_index()

pizza_cat.sort_values(by = "sales", ascending = False).reset_index(drop = True)
```

This shows the categories of pizzas based on revenue

```python
#Pizza size that generates the highest revenue

pizza_size = df.groupby("size")["sales"].sum().reset_index()

pizza_size.sort_values(by="sales", ascending=False)
```

This shows the pizza size that generated most revenue and it is observed that large(L) has the highest revenue

```python
#Top 10 ordered pizza by quantity

pizza_by_quant = df.groupby("name")["quantity"].sum().reset_index()

pizza_by_quant.sort_values(by="quantity", ascending=False).head(10).reset_index(drop=True)
```

Based on quantity, The Classic Deluxe Pizza has the highest quantity being ordered

```python
#Most popular size

pop_size_quant = df.groupby("size")["quantity"].sum().reset_index()

pop_size_quant.sort_values(by="quantity", ascending=False)
```

This shows the most popular size and the result shows that large(L) has the most orders

```python
#For each pizza, what size do people order most

most_pop_order_size = df.groupby(["name","size"])["order_id"].count().reset_index()

most_pop_order_size.sort_values(by="order_id", ascending=False).reset_index(drop=True)
```

This shows the size of pizza people order the most on each pizza

```python
total_mean_qty = df.groupby("name").agg(total_sales = ("sales", "sum"), avg_price = ("price","mean"), total_qty = ("quantity","sum")).reset_index()

total_mean_qty.sort_values(by = ["total_sales", "avg_price", "total_qty"], ascending = [False, False, False]).reset_index(drop=True)
```

The above shows the total sales, average price and total quantity from each pizza and it is observerd that though The Thai Chicken Pizza has the highest sales, The Classic Deluxe Pizza has most quantity sold with lesser revenue generated compared to The Thai Chicken Pizza. The Brie Carre Pizza has the highest average price per unit but has the least sales and least quantity sold in the Pizza Place
